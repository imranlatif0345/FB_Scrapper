```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
load_dotenv
mail = os.getenv("ACCT_ID")
acct_pass = os.getenv("ACCT_PWD")
# Set up WebDriver
driver = webdriver.Chrome()

# Log in to Facebook
driver.get('https://www.facebook.com')
email = driver.find_element(By.ID, 'email')
password = driver.find_element(By.ID, 'pass')
email.send_keys(mail)
password.send_keys(acct_pass)
password.send_keys(Keys.RETURN)

# Wait for the login process to complete
# WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '//div[@role="feed"]')))
time.sleep(5)
# Navigate to the Group
group_url = 'https://www.facebook.com/groups/643777109561565/'
driver.get(group_url)

# Wait for the group page to load
WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '//div[@role="feed"]')))

# Scroll to Load More Posts (if needed)
def scroll_down():
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Wait for the page to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

scroll_down()

# Scrape Posts
posts = driver.find_elements(By.XPATH, '//div[@role="feed"]/div')

for post in posts:
    try:
        # Extract creator name
        creator = post.find_element(By.XPATH, './/h2').text
        
        # Extract time of the post
        # time_element = post.find_element(By.XPATH, './/abbr')
        time_element = post.find_element(By.XPATH, './/span[contains(@class, "timestampContent")]')

        post_time = time_element.get_attribute('title')
        
        # Extract post content (optional)
        content = post.find_element(By.XPATH, './/div[@data-ad-preview="message"]').text
        
        print(f'Creator: {creator}\nTime: {post_time}\nContent: {content}\n')
    
    except Exception as e:
        print(f"An error occurred: {e}")

# Close the WebDriver
driver.quit()
```
